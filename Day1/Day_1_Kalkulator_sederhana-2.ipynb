```python
def cek_status():
    while True :
        aktif = input("Apakah aktif (ya/tidak) ?")
        if aktif.lower() == 'ya':
            return True
        elif aktif.lower() == 'tidak':
            return False
        else :
            print("Input tidak valid. Silahkan coba lagi")

user_data = []
while True:
  nama = input( "Nama :")
  umur = int(input("Umur :"))
  tinggi = float(input("Tinggi :"))
  aktif = cek_status()


  data = {
      "Nama": nama,
      "Umur": umur,
      "Tinggi": tinggi,
      "Aktivitas": aktif
  }

  user_data.append(data)
  end = input("Ketik (selesai) jika sudah selesai :")
  if end.lower() == 'selesai':
      break

print("\n-- Data telah tersimpan ---")
for i, user in enumerate(user_data, 1):
    print(f"{i}. Nama: {user['Nama']} | Umur : {user['Umur']} | Tinggi :{user['Tinggi']:.0f}cm | Aktif: {'Ya' if user['Aktivitas'] else 'Tidak'} " )
```

```python
def kalkulator (num1, num2):
    print(f"Penjumlahan   : {num1} + {num2} = {num1 + num2}")
    print(f"Pengurangan   : {num1} - {num2} = {num1 - num2}")
    print(f"Perkalian     : {num1} * {num2} = {num1 * num2}")
    if num2 != 0:
      print(f"Pembagian     : {num1} / {num2} = {num1 / num2 :.2f}")
      print(f"Sisa bagi     : {num1} % {num2} = {num1 % num2}")
    else:
      print("Tidak dapat melakukan pembagian dengan nol.")
    print(f"Pangkat: {num1} ^ {num2} = {num1 ** num2}")

def main():
  num1 = float(input("Masukkan angka pertama:"))
  num2 = float(input("Masukkan angka kedua:"))
  hasil = kalkulator(num1, num2)

main()
```

```python
def kalkulator_interaktif (num1, ops, num2):
    if ops == "+":
        return num1 + num2
    elif ops == "-":
        return num1 - num2
    elif ops == "*":
        return num1 * num2
    elif ops == "/":
        if num2 != 0:
          return num1 / num2
        else:
          print("tidak bisa membagi dengan 0")
          return None
    elif ops == "%":
        return num1 % num2
    elif ops == "^":
        return num1 ** num2
    else :
        print("Operator tidak valid")

def main():
  while True :
      num1 = float(input("masukkan angka pertama:"))
      ops = input("masukkan operator (+,-,*,/,%,^):")
      num2 = float(input("masukkan angka kedua :"))

      hasil = kalkulator_interaktif(num1,ops,num2)
      if hasil is not None:
          print(f"Hasil kalkulasi: {num1} {ops} {num2} = {hasil}")


      lagi = input("Hitung lagi? (y/n) :")
      if lagi.lower() == 'n':
          print("Terimakasih")
          break

main()
```
